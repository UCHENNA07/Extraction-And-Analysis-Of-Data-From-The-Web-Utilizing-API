# ANALYSING TESLA DATA

```python
STOCK_NAME = "TSLA"
COMPANY_NAME = "Tesla Inc"

STOCK_ENDPOINT = "https://www.alphavantage.co/query"
NEWS_ENDPOINT = "https://newsapi.org/v2/everything"

STOCK_API_KEY = 'P7FEEGDXXCOBSTM'
NEWS_API_KEY = '1c5a322ed48b46e39bae5f45f649dffb'
```

```python
stock_params = {
    'function': 'TIME_SERIES_DAILY',
    'symbol': STOCK_NAME,
    'apikey': STOCK_API_KEY
}
```

```python
response = requests.get(STOCK_ENDPOINT).json()
response
```

The above error message indicates that 'the parameter apikey is invalid or missing'. When extracting data from the web, some API sources requires APIKEY to successfully get the data. Other parameters are included so as to get the accurate format of the particular data you are interested in. Below is the correct format.

```python
response = requests.get(STOCK_ENDPOINT, params=stock_params)
```

```python
data = response.json()['Time Series (Daily)']
```

```python
data
```

Note: This data updates on a daily basis and updates subsequently

```python
data_list = [value for (key, value) in data.items()]
yesterday_data = data_list[0]
yesterday_data
```

```python
yesterday_closing_price = yesterday_data['4. close']
yesterday_closing_price
```

```python
day_before_yesterday_data = data_list[1]
day_before_yesterday_data
```

```python
day_before_yesterday_closing_price = day_before_yesterday_data['4. close']
day_before_yesterday_closing_price
```

```python
diff = abs(float(yesterday_closing_price) - float(day_before_yesterday_closing_price))
diff
```

```python
if diff > 2:
    news_params = {
    'apiKey': NEWS_API_KEY,
    'qInTitle': COMPANY_NAME,
}
new_response = requests.get(NEWS_ENDPOINT, params=news_params)
article = new_response.json()['articles']
message = article[:1]
message
```
